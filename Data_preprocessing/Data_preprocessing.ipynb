### Лемматизация

```python
from pymystem3 import Mystem
m = Mystem()
```

```python
from collections import Counter
```

```python
data = df['purpose']
```

```python
data
```

Создал переменную для столбца с значениями purpose

```python
def lammatiz(data):
    count = []
    for row in data:
        lemmas = m.lemmatize(row)
        print(lemmas)
        count += lemmas
    return count
```

Функция для применения m.lemmatize()

```python
count = lammatiz(data)
```

Сохранил в отдельную переменную
